```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn import tree
```

```python
df = pd.read_csv("play_tennis.csv")
```

```python
data = df
data
```

```python
le_out = LabelEncoder()
le_temp =LabelEncoder()
le_hum=LabelEncoder()
le_wind=LabelEncoder()
```

```python
data["outlook_t"] = le_out.fit_transform(data["outlook"])
data["temp_t"] = le_temp.fit_transform(data["temp"])
data["humidity_t"] = le_hum.fit_transform(data["humidity"])
data["wind_t"] = le_wind.fit_transform(data["wind"])
```

```python
data.to_csv("play_bv.csv")
```

```python
data = df.drop(["humidity","temp","outlook","wind",'day'],axis = 'columns')
data
```

```python
x = data[['outlook_t','temp_t','humidity_t','wind_t']]
```

```python
y = data["play"]
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state = 10)
```

```python
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[1,2,0,2]])
```

```python
model.predict(X_test)
```

## Exercise: Build decision tree model to predict survival based on certain parameters

```python
url = "https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/9_decision_tree/Exercise/titanic.csv"
f = pd.read_csv(url)
```

```python
f.columns
```

```python
dt = f.drop(['PassengerId', 'Name','SibSp','Ticket','Parch', 'Cabin', 'Embarked',],axis="columns")
dt
```

```python
lc = LabelEncoder() 
```

```python
dt["sex_t"] = lc.fit_transform(dt["Sex"])
```

```python
dt.head(3) #male = 1, female= 0
```

```python
dt = dt.drop("Sex",axis ="columns")
```

```python
dt.head(3)
```

```python
m = dt["Age"].median()
m
```

```python
dt["Age"].fillna(m,inplace = True)
```

```python
dt.isna().sum()
```

```python
x = dt[['Pclass','sex_t','Age','Fare']]
```

```python
y = dt["Survived"]
```

```python
t_model = tree.DecisionTreeClassifier()
```

```python
 X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
t_model.fit(X_train,y_train)
```

```python
t_model.predict(X_test)
```

```python
t_model.score(X_test,y_test)
```
