```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices_3.csv")
```

```python
df
```

```python
dummy = pd.get_dummies(df.town,dtype=int)
dummy 
```

```python
new_df = pd.concat([df,dummy],axis ="columns")
```

```python
new_df = new_df.drop(["west windsor"],axis = "columns")
new_df
```

```python
x = new_df.drop(["price"],axis="columns")
x
```

```python
y = new_df["price"]
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[3400,0,0]])
```

```python
model.predict([[2800,0,1]]) 
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("carprices.csv")
```

```python
df
```

```python
dummy = pd.get_dummies(df["Car Model"],dtype = int)
dummy
```

```python
new_df = pd.concat([df,dummy],axis="columns")
```

```python
new_df = new_df.drop(["Car Model"],axis = "columns")
```

```python
new_df
```

```python
x = new_df.drop(["Sell Price($)"],axis = "columns")
```

```python
y = new_df["Sell Price($)"]
y
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```
