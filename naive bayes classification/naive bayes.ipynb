```python
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head(5)
```

```python
inputs = df.drop(["PassengerId","Name","Cabin","Embarked","Ticket","SibSp","Parch"],axis="columns")
```

```python
inputs.head()
```

```python
target = inputs.Survived
inputs.drop("Survived",axis="columns",inplace=True)
target
```

```python
inputs.head(5)
```

```python
dummies = pd.get_dummies(inputs.Sex,dtype=int)
dummies
```

```python
inputs = pd.concat([inputs,dummies],axis="columns")
```

```python
inputs.drop("Sex",axis="columns",inplace= True)
```

```python
inputs
```

```python
inputs.isna().sum()
```

```python
import math
mean = (inputs.Age.mean())
inputs.Age = inputs.Age.fillna(mean)
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.20)
```

```python
model = GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
model.predict(X_test)
```
